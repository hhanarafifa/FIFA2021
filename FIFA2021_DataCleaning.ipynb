```python
# Look at the overview of the weight data
# They are not in the same units. Some are in kg, while some are in lbs
# Thus, we neet to convert it all to kg

df['Weight'].unique()
```

```python
# Create function to convert the weight

weights_in_kg = []

for weight in df.Weight:
    if "kg" in weight:
        weight =  weight.strip('kg')
        weights_in_kg.append(weight)
    elif "lbs" in weight:
        weight = float(weight.strip('lbs'))
        weight = weight/2.205
        weights_in_kg.append(weight)
```

```python
# apply the function and convert it to integer
df['Weight'] = pd.Series(weights_in_kg, dtype = float)
df['Weight'] = df['Weight'].astype(int)
df['Weight'].unique()
```

```python
# Extract only number from W/F, SM, IR
df["W/F"] = df["W/F"].str.replace('\D+', '')
df["SM"] = df["SM"].str.replace('\D+', '')
df["IR"] = df["IR"].str.replace('\D+', '')

# Change the datatype of those columns
df[['W/F','SM','IR']]= df[['W/F','SM','IR']].astype(int)

# df
```

```python
# removing any new line characters
df = df.replace('\n',' ', regex=True)
# df
```

```python
df[['Value','Wage','Release Clause']] = df[['Value','Wage','Release Clause']].replace('€','', regex = True)
# df
```

```python
# Convert K to 1000, M to 1000000, and B to 1000000000

def string_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df['Value'] = df['Value'].apply(string_to_float)
df['Wage'] = df['Wage'].apply(string_to_float)
df['Release Clause'] = df['Release Clause'].apply(string_to_float)

df['Hits'] = df['Hits'].fillna(0)
df['Hits'] = df['Hits'].apply(string_to_float)

# Change the datatype of those columns
df[['Value','Wage','Release Clause', 'Hits']]= df[['Value','Wage','Release Clause', 'Hits']].astype(int)
```

```python
# Change Joined Loan Date End column datatype into Date Time Data
df[["Joined","Loan Date End"]]=df[["Joined","Loan Date End"]].apply(pd.to_datetime)
# df.info()
```

```python
# From the contract column, create start_contract and end_contract
# Joined 

df['start_contract'] = df['Contract'].str[:4]
df['end_contract'] = df['Contract'].str[7:]

df['start_contract']=pd.to_datetime(df['start_contract'], errors = 'coerce')
df['end_contract']=pd.to_datetime(df['end_contract'], errors = 'coerce')
df
```

```python
# Calculate contract duration from end_contract and start_contract column
df['contract_duration'] = round((df['end_contract']-df['start_contract'])/np.timedelta64(1,'Y'),0)

df['contract_duration'] = df['contract_duration'].fillna(0)
df['contract_duration'] = df['contract_duration'].astype(int)

# df
```

```python
# Extract player name from the playerUrl

df['player_name'] = df['playerUrl'].str.split('/').str[-3].str.split('/').str[0]

# Replace the '-' and make it a title case
df["player_name"] = df["player_name"].str.replace('-', ' ').str.title()
# df
```
