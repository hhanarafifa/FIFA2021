## All column names are in the same format! Let's start the data preprocessing

```python
# Drop player ID
df = df.drop(['id','longname'], axis = 1)
```

```python
# Check for unique character
def unique_characters( df, col ):
    result = []
    for index, row in df.iterrows():
        for char in row[col]:
            if char not in result:
                result.append(char)
    return result
```

```python
print(unique_characters(df,'name'))
print(unique_characters(df,'club'))
print(unique_characters(df,'nationality'))
```

```python
from unidecode import unidecode
df['club'] = df['club'].apply(unidecode)
df['name'] = df['name'].apply(unidecode)
df['nationality'] = df['nationality'].apply(unidecode)

df['club'] = df['club'].str.encode('ascii', 'ignore').str.decode('ascii')
df['name'] = df['name'].str.encode('ascii', 'ignore').str.decode('ascii')
df['nationality'] = df['nationality'].str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
print(unique_characters(df,'name'))
print(unique_characters(df,'club'))
print(unique_characters(df,'nationality'))
```

```python
# Look at the overview of the weight data
# They are not in the same units. Some are in kg, while some are in lbs
# Thus, we neet to convert it all to kg

df['weight'].unique()
```

```python
# Create function to convert the weight

weights_in_kg = []

for weight in df.weight:
    if "kg" in weight:
        weight =  weight.strip('kg')
        weights_in_kg.append(weight)
    elif "lbs" in weight:
        weight = float(weight.strip('lbs'))
        weight = weight/2.205
        weights_in_kg.append(weight)
```

```python
# apply the function and convert it to integer
df['weight'] = pd.Series(weights_in_kg, dtype = float)
df['weight'] = df['weight'].astype(int)
```

```python
# value in value, wage, and release_clause contains € and letter like 'K','M','B'.
# Remove € and convert 'K' into 1000, 'M' into 1000000, and 'B' into '1000000000'
```

```python
# remove € symbol
df[['value','wage','release_clause']] = df[['value','wage','release_clause']].replace('€','', regex = True)

# create function to convert letters to numbers
def string_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    else:
        return x

df['value'] = df['value'].apply(string_to_float)
df['wage'] = df['wage'].apply(string_to_float)
df['release_clause'] = df['release_clause'].apply(string_to_float)

df['hits'] = df['hits'].apply(string_to_float)
```
